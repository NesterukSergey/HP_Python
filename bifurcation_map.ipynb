```
import random
import matplotlib.pyplot as plt
import numpy as np

from pylab import rcParams
rcParams['figure.figsize'] = 15, 10
```

```
def bifurcation_step(previous_value, r):
  return r * previous_value * (1 - previous_value)
```

```
def bifurcation_points(x, r, loops, skip):
  # x = random.random()

  values = [x]
  for i in range(loops + skip):
    values.append(bifurcation_step(values[i], r))

  return values[skip:]
```

```
iterations = 100
r = 3
x = random.random()
x
```

```
plt.plot(bifurcation_points(x, r,loops=200, skip=200))
plt.xlabel('Iteration')
plt.ylabel('x')
```

```
def draw_bifurcation_map(min_r=0.5, max_r=4, loops=200, skip=200):
  r_list = np.linspace(min_r, max_r, 250)
  x = random.random()

  R = []
  y = []

  for r in r_list:
    for p in bifurcation_points(x, r, loops, skip):
      R.append(r)
      y.append(p)
  
  plt.scatter(R, y, s=1)
  plt.xlabel('r')
  plt.ylabel('x')
```

##Bifurcation map

```
draw_bifurcation_map(min_r=2.5, max_r=5, loops=1000, skip=1000)
```
